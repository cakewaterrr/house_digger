```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
location = 'Corpus-Christi'
listing_type = 'sale' # 'rent', 'sale'
property_type = ['house', 'condo', 'townhouse', 'land', 'multifamily'] 
min_price = '100k'
max_price = '1M'
min_beds = '1'
max_beds = '4'
min_baths = '1.5'

if listing_type == 'rent':
    url = f'https://www.redfin.com/city/35781/TX/Corpus-Christi/apartments-for-rent/filter/property-type={property_type[0]},min-price={min_price}k,max-price={max_price}k,min-beds={min_beds},min-baths={min_baths}'
else:
    url = f'https://www.redfin.com/city/35781/TX/Corpus-Christi/filter/property-type={property_type[0]},min-price={min_price}k,max-price={max_price}k,min-beds={min_beds},max-beds={max_beds}min-baths={min_baths}'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
homes = {'Address': [],
         'Price': [],
         'Beds': [],
         'Baths': [],
         'Footage': []}

for i in range(20):
    home_containers = soup.find_all(class_='HomeCardContainer', id=f'MapHomeCard_{i}')
    stats=home_containers[0].find_all(class_='stats')

    homes['Price'] += [(home_containers[0].find(class_='homecardV2Price').get_text())]
    homes['Address'] += [home_containers[0].find(class_='homeAddressV2').get_text()]
    homes['Beds'] += [stats[0].get_text()]
    homes['Baths'] += [stats[1].get_text()]
    homes['Footage'] += [stats[2].get_text()]



properties = pd.DataFrame.from_dict(homes)
```
