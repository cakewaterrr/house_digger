```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

page = requests.get("https://www.redfin.com/city/35781/TX/Corpus-Christi/filter/sort=lo-price,property-type=house,min-price=150k,max-price=900k,min-beds=1,max-beds=3,min-baths=2,status=active", headers = {'User-agent': 'Super Bot Power Level Over 9000'})

if page.status_code == 200:
    soup = BeautifulSoup(page.text, 'html.parser')

    home_containers = soup.find_all(id='content')
else: 
    print("page status code:" + str(page.status_code))

home_containers
```

```python
location = 'Corpus-Christi'
listing_type = 'sale' # 'rent', 'sale'
property_type = ['house', 'condo', 'townhouse', 'land', 'multifamily'] 
min_price = '100k'
max_price = '1M'
min_beds = '1'
max_beds = '4'
min_baths = '1.5'

if listing_type == 'rent':
    url = f'https://www.redfin.com/city/35781/TX/Corpus-Christi/apartments-for-rent/filter/property-type={property_type[0]},min-price={min_price}k,max-price={max_price}k,min-beds={min_beds},min-baths={min_baths}'
else:
    url = f'https://www.redfin.com/city/35781/TX/Corpus-Christi/filter/property-type={property_type[0]},min-price={min_price},max-price={max_price},min-beds={min_beds},max-beds={max_beds}min-baths={min_baths}'
    
```

```python
homes = {'Address': [],
         'Price': [],
         'Beds': [],
         'Baths': [],
         'Footage': []}

for i in range(20):
    home_containers = soup.find_all(class_='HomeCardContainer', id=f'MapHomeCard_{i}')
    stats=home_containers[0].find_all(class_='stats')

    homes['Price'] += [(home_containers[0].find(class_='homecardV2Price').get_text())]
    homes['Address'] += [home_containers[0].find(class_='homeAddressV2').get_text()]
    homes['Beds'] += [stats[0].get_text()]
    homes['Baths'] += [stats[1].get_text()]
    homes['Footage'] += [stats[2].get_text()]



properties = pd.DataFrame.from_dict(homes)
```
